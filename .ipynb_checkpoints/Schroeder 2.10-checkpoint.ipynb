#### Problem 2.10 

It is given in the problem that we have a Einstein solid of 200 oscillators, 
We can use the multiplicity expression:

$$ \Omega (N,q) = { (q+N-1) \over {q!(N-1)!}}$$

So since we have two solids of different harmonic oscillators, $ N $ we can express the total multiplicity as:

$ \Omega_{total} (N,q) = \Omega (N,q_A) * \Omega (N,q_B) =  { (q_A+N_A-1) \over {q_A!(N_A-1)!}}  { (q_B+N_B-1) \over {q_B!(N_B-1)!}} $

Where A is the first Einstein solid, and B is the second.

From the above equation we can construct a table:

```python
import pandas as pd
import numpy as np
import scipy as sp
import math   
```

```python
def mcity(q,n):
    return (math.factorial(q+n-1))/(math.factorial(q)*math.factorial(n-1))
```

```python
qA= [i+1 for i in xrange(100)]
qB = [100-i for i in xrange(100)]
omegaA=[float(mcity(i,200)) for i in xrange(100)]
omegaB=[float(mcity(100-i,100)) for i in xrange(100)]
omegaTotal=[a*b for a,b in zip(omegaA,omegaB)]
```

```python
data = {'$q_A$':qA, 
        '$\Omega (N,q_A)$':omegaA,
        '$q_B$':qB,
        '$\Omega (N,q_B)$':omegaB,
        '$\Omega_{total}$':omegaTotal}

df = pd.DataFrame(data,columns=['$q_A$','$\Omega (N,q_A)$','$q_B$','$\Omega (N,q_B)$', '$\Omega_{total}$'])
df
```

Now if we plot these results between $ q_A $ and $ \Omega_{total} $

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(qA,omegaTotal)
plt.show()
```

We are able to find the most probable macrostate:

```python
maxvalue = max(omegaTotal)
print omegaTotal.index(maxvalue)
print 100-omegaTotal.index(maxvalue)
```

$ P = { \Omega(n)\over \Omega(all) } $

```python
omegaTotal[67]/sum(omegaTotal)
```

so the probability is $ P =7.3 $%

```python
minvalue = min(omegaTotal)
print omegaTotal.index(minvalue)
print 100 - omegaTotal.index(minvalue)
```
